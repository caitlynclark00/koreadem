```python
import numpy as np
import pandas as pd
from pandas import *
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import matplotlib.ticker as ticker
plt.rcParams['font.family'] = 'Times New Roman'
```

```python
data = read_csv("Organizing Rate - Sheet1.csv")
data.set_index("YEAR", inplace=True)
plot = data.plot()

plot.lines[0].set_linestyle(':') 
plot.lines[1].set_linestyle('--') 
plot.lines[2].set_linestyle('-')

plot.set_xticks([1970, 1972, 1974, 1976, 1978, 1980, 1982, 1984, 1986, 1988, 1990])
plot.set_title("Labor Organizing Rate Over Time, South Korea")
plot.set_xlabel("Year")
plot.set_ylabel("Organizing Rate (%)")
plot.legend(['Men', 'Women', 'Total'])

# Add shading
plot.axvspan(1987, 1990, facecolor='whitesmoke', edgecolor='none')

# Add box around democratization label
bbox_props = dict(boxstyle='square,pad=0.3', fc='white', ec='black', lw=1)
plt.text(1986.8, 26, 'Democratization', va='top', ha='right', bbox=bbox_props)


plt.savefig("organizing_rate.png", dpi=300)
plt.show()
```

```python
korea_number = read_csv("kor strike number.csv")
korea_renamed = korea_number.rename(columns={"time": "year"})
korea_pivot = korea_renamed.pivot(index = 'year', columns = 'classif1.label', values = 'obs_value')
korea_pivot = korea_pivot.drop(['Economic activity (Aggregate): Total', 'Economic activity (Broad sector): Total'
, 'Economic activity (ISIC-Rev.2): 0. Activities not Adequately Defined'], axis = 1)

korea_plot = korea_pivot.plot()
korea_plot.lines[-1].set_linestyle("--")
korea_plot.legend(['Agriculture', 'Mining', 'Manufacturing', 'Electricity, Gas & Water',
'Construction', 'Retail & Service', 'Transportation & Logistics', 'Finance & Real Estate', 'Social Services', 'Total'])
plt.xlabel('Year')
plt.ylabel('Number of Strikes')
korea_plot.set_title('Number of Strikes by Sector, South Korea')
korea_plot.set_xticks([1970, 1972, 1974, 1976, 1978, 1980, 1982, 1984, 1986, 1988, 1990])


plt.xlim(1970, 1990)
plt.ylim(0, 5000)

# shading the right side after 1987
korea_plot.add_patch(Rectangle((1987, 0), 1990 - 1987, 5000, facecolor='whitesmoke', edgecolor='none'))


bbox_props = dict(boxstyle="square,pad=0.3", fc="white", ec="black", lw=1)
plt.text(1986.8, 4500, 'Democratization', va='top', ha='right', bbox=bbox_props)

plt.savefig("strike_num.png", dpi=300)
plt.show()
```

```python
#Figure 3
korea_days = read_csv("korea strike days.csv")
korea_days = korea_days.pivot(index = 'time', columns = 'classif1.label', values = 'obs_value')
korea_days = korea_days.drop(['Economic activity (Aggregate): Total', 'Economic activity (Broad sector): Total',
'Economic activity (ISIC-Rev.2): 0. Activities not Adequately Defined'], axis = 1)
days_plt = korea_days.plot()
days_plt.lines[-1].set_linestyle("--")
plt.legend(['Agriculture', 'Mining', 'Manufacturing', 'Electricity, Gas & Water',
'Construction', 'Retail & Service', 'Transportation & Logistics', 'Finance & Real Estate', 'Social Services', 'Total'])
plt.xlabel('Year')
plt.ylabel('Days on Strike')

formatter = ticker.FuncFormatter(lambda x, p: format(int(x), ','))
plt.gca().yaxis.set_major_formatter(formatter)

plt.title('Days Not Worked Due to Strikes & Lockouts by Sector, South Korea')
plt.xticks([1970, 1972, 1974, 1976, 1978, 1980, 1982, 1984, 1986, 1988, 1990])

days_plt.add_patch(Rectangle((1987, 0), 1990 - 1987, 10000000000000000000, facecolor='whitesmoke', edgecolor='none'))


bbox_props = dict(boxstyle="square,pad=0.3", fc="white", ec="black", lw=1)
plt.text(1986.8, 6500000, 'Democratization', va='top', ha='right', bbox=bbox_props)
plt.savefig("days_strike.png", dpi=300)
plt.show()
```
